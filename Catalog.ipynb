# Description

This notebook is used to request data information of WaPOR database using WaPOR API.

```python
import requests
import json
import pandas as pd

url = 'https://io.apps.fao.org/gismgr/api/v1/catalog/workspaces?overview=false&paged=false'
```

# Get list of workspaces

```python
resp=requests.get(url).json()
workspaces=pd.DataFrame.from_dict(resp['response'])
workspaces
```

# Get list of data cubes in a workspace

```python
workspace='WAPOR'
url_cube= 'https://io.apps.fao.org/gismgr/api/v1/catalog/workspaces/{0}/cubes?overview=false&paged=false'.format(workspace)
resp=requests.get(url_cube).json()
resp
catalog=pd.DataFrame.from_dict(resp['response'])
catalog
```
